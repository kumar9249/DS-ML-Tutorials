# Vizualization using Seaborn

## Seaborn Tutorials

### Relational Plots:

relplot() - Figure-level interface for drawing relational plots onto a FaceGrid

scatterplot() - Draw a scatterplot with possibility of several semantic groupings

lineplot() - Draw a line plot with possibility of several semantic groupings

### Categorical Plots:

catplot() - Figure-level interface for drawing categorical plots onto a FaceGrid


-> Categorical scatterplots

stripplot() - Draw a scatterplot where one variable is categorical

swarmplot() - Draw a categorical scatterplot with non-overlapping points


-> Categorical distribution plots

boxplot() - Draw a boxplot to show distributions with respect to categories

violinplot() - Draw a combination of boxplot and kernel density estimate

boxenplot() - Draw an enhanced boxplot for larger datasets 


-> Categorical estimate plots

pointplot() - Show point estimates and confidence intervals using scatterplot 

barplot() - Show point estimates and confidence intervals as rectangular bars

countplot() - Show the counts of observations in each categorical bins using bars

### Distribution Plots:

distplot() - Flexibly plot a univariate distribution of observations

kdeplot() - Fit and plot a univariate or bivariate kernel density estimate

rugplot() - Plot datapoints in an array as sticks on an axis

### Regression Plots:

lmplot() - Plot data and regression model fits across a FaceGrid

regplot() - Plot data and a linear regression model fit

residplot() - Plot the residuals of a linear regression

### Matrix Plots:

heatmap() - Plot rectangular data as a color-encoded matrix

clustermap() - Plot a matrix dataset as a hierarchically-clustered heatmap

### Bonus Plots:

jointplot() - Plot paiwise realtionships in a dataset 

pairplot() - Draw a plot of two variables with bivariate and univariate graphs

```python
# Violin Plot - violinplot()

# x, y, hue - Inputs for plotting long-form data
# data - Dataset for plotting. If x and y are absent, this is interpreted as wide-form
# order, hue_order - Order to plot the categorial levels in
# orient - {"v"/"h"} Orientation of the plot
# dodge - Seperate the strips for different hue levels along the categorial axis
# palette -
# linewidth -
# saturation -
# width - 
# bw - {'scott', 'silverman', float} The scale factor to use when computing the kernel bandwidth
        # The actual kernel size will be determined by multipying the scale factor by the standard deviation of the data within each bin
# cut - Distance, in unit of bandwidth size, to extend the density past the extreme datapoints
        # Set to 0 to limit the violin range within the range of the observed data
# scale - {'area', 'count', 'width'}
        # If area, each violin will have the same area
        # If count, the width of the violins will be scaled by the no. of observations in that bin.
        # If width, each violin will have the same width
# inner - {'box', 'quartile', 'point', 'stick', None} Representation of the datapoints in the violin interior
        # If box, draw a miniature boxplot
        # If quartile, draw the quartiles of the distribution
        # If point or stick, show the each underlying datapoint
# gridsize - No. of points in the discrete grid used to compute the kernel density estimate
# split - Draw half of a violin for each hue level
```

```python
# Regression Plot - regplot()

# x, y - Input variables
# data - Long-form dataframe
# x_estimator - Apply this function to each unique value if x and plot the resulting estimate
              # This is useful when x is a discrete variable
# x_bins - Bin the x variable into discrete bins and then estimate the central tendency and a confidence interval
         # This binning only influences how the scatterplot is drawn
         # When this parameter is used, it implies that the default of x_estimator is np.mean
# x_ci - {'ci', 'sd', int, None} Size of the confidence interval used when plotting a central tendency for discrete values of x
# scatter - If true, draw a scatterplot with the underlying observations
# fig_reg - If true, estimate and plot a regression model relating the x and y variables 
# ci - Size of the confidence intervalfor the regression estimate
# order - 
# logistic - If true, assume that y is a binary variable and use statsmodel to estimate a logistic regression model; Set ci to None
# lowess - If true, use statsmodel to estimate a nonparametric locally weighted linear regression
# robust - If true, use statsmodel to estimate a robust regression. This will de-weight outliers; Set ci to None
# log_x - If true, estimate a linear regression of the form y~log(x); x must be positive 
# {x,y}_partial - Confounding variables to regress out of the x or y variables before plotting 
# truncate - If true, the regression line is bounded by the data limits otherwise it extends to the x axis limits
# {x,y}_jitter - Add a uniform noise of this size to either the x or y variables 
# label - 
# color - Color to apply to all plot elements
# marker - Marker to use for the scatterplot
```

```python
# Heatmap - heatmap()

# data - 
# vmin, vmax - Values to anchor the colormap
# cmap -
# center - The value at which to center the colormap when plotting divergent data
# robust - 
# annot - If true, write the data value in each cell
# fmt - String formatting code to use when adding annotations
# annot_kws -
# linewidths - Width of the lines that will divide each cell
# linecolor - Color of the lines that will divide each cell
# cbar -
# cbar_kws -
# cbar_ax - Axes in which to draw the colorbar
# square - If true, set the Axes aspect to "equal" so each cell will be square-shaped
# xticklabels, yticklabels - If true, plot the column names of the dataframe
# mask - If passed, data will not shown in cells where mask is true
```

```python
np.random.seed(0)
sns.set()
uniform_data = np.random.randn(10, 12)
ax = sns.heatmap(uniform_data, vmin=0, vmax=1, yticklabels=False, square=True)
```

```python
normal_data = np.random.randn(10, 12)
ax = sns.heatmap(normal_data, center=0, linecolor='k')
```

```python
flights = sns.load_dataset("flights")
flights = flights.pivot("month", "year", "passengers")
ax = sns.heatmap(flights)
```

```python
ax = sns.heatmap(flights, annot=True, fmt='d')
```

```python
ax = sns.heatmap(flights, center=flights.loc["January", 1955], cmap="YlGnBu", linewidths=.5, cbar=False)
```

```python
# Pair Plot - pairplot()

# data - Long-form dataframe
# hue - Variable in data to map plot aspects to different colors
# hue_order - Order for hue levels
# palette -
# vars - Variables within data to use
# {x,y}_vars - Variables within data to use seperately for the rows and columns 
# kind - {'scatter', 'reg'} Kind of plot for non-identity relationships
# diag_kind - {'auto', 'hist', 'kde', None} Kind of plot for the diagonalsubplots
# markers -
# height -
# aspect -
# corner -
# dropna - Drop missing values from the data before plotting
# {plot,diag,grid}_kws -
```

```python
sns.set(style="ticks", color_codes=True)
g = sns.pairplot(iris)
```

```python
g = sns.pairplot(iris, hue="species")
```

```python
g = sns.pairplot(iris, vars=["sepal_width", "sepal_length"])
```

```python
g = sns.pairplot(iris, x_vars=["sepal_width", "sepal_length"], y_vars=["petal_width", "petal_length"])
```

```python
g = sns.pairplot(iris, kind="reg", diag_kind="kde", height=2, aspect=1.2)
```

```python
# Joint Plot - jointplot()

# x,y -
# data -
# kind - {'scatter', 'reg', 'resid', 'kde', 'hex'} Kind pf plot to draw
# color - Color used for the plot elements
# height - Size of the figure (it will be square)
# ratio - Ratio of joint axes height to marginal axes height
# space - Space between the joint and marginal axes
# dropna - If true, remove observations that are missing from x and y 
# {x,y}_lim - Axis limits to set before plotting
# {joint,marginal,annot}_kws - 
```

```python
sns.set(style="white", color_codes=True)
g = sns.jointplot(x="total_bill", y="tip", data=tips, height=6.5, ratio=3, space=0.3)
```

```python
g = sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg", color='r')
```

```python
g = sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex", color='r')
```

```python
g = sns.jointplot(x="sepal_width", y="petal_length", data=iris, kind="kde", color='g')
```

```python
g = sns.jointplot(x="total_bill", y="tip", data=tips, kind="resid")
```
