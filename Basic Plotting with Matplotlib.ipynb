```python
# Subplots

import numpy as np
from numpy.random import randn
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(5, 3))

fig.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=None)

# Specifying padding & spacing around the subplots
# wspace and hspace controls the percent of the figure width and figure height

# plt.subplot(nrows, ncols, index)
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)

# fig, ax = plt.subplots(nrows= , ncols= , sharex= , sharey= )

# nrows - Number of rows of subplots
# ncols - Number of columns of subplots
# sharex - All subplots should use the same X-axis ticks (adjusting the xlim will affect all subplots)
# sharey - All subplots should use the same Y-axis ticks (adjusting the ylim will affect all subplots)
# sharex & sharey - ['row', 'col', 'all', 'none']

ax1.hist(randn(100), bins=20, color='k', alpha=0.3)
ax2.scatter(np.arange(30), np.arange(30) + 3 * randn(30))
ax3.plot(randn(50).cumsum(), 'k--')

plt.show()
```

```python
# plt.subplot2grid((R, C), loc, rowspan= , colspan= )

T = np.linspace(-np.pi, np.pi, 1024)

grid_size = (4, 2)

plt.subplot2grid(grid_size, (0, 0), rowspan = 3, colspan = 1)
plt.plot(np.sin(2 * T), np.cos(0.5 * T), c = 'k')

plt.subplot2grid(grid_size, (0, 1), rowspan = 3, colspan = 1)
plt.plot(np.cos(3 * T), np.sin(T), c = 'k')

plt.subplot2grid(grid_size, (3, 0), rowspan=1, colspan=3)
plt.plot(np.cos(5 * T), np.sin(7 * T), c= 'k')

plt.tight_layout()
plt.show()
```

```python
# Histograms 

# It is a bar chart which shows the frequency for a given phenomena. eg. Probability Distribution
# Probability density functions - x-axis: probability a given value would occur (0 t0 1) 
# X-axis is a given observation and Y-axis is the frequency with which that observation occurs

X = np.random.randn(1000)
plt.hist(X, bins = [0, 0.5, 1, 1.5, 2, 2.5, 3], alpha=0.5, histtype='stepfilled', color='b', edgecolor='k')
plt.show()

# By default, the range of the values will be divided into 10 equal-sized bins
# n = No. of observations
# Range = max value - min value
# No. of intervals = (n)^1/2
# Width of intervals = Range / No. of intervals

# histtype - ['bar', 'barstacked', 'step', 'stepfilled']
# orientation - ['vertical', 'horizontal']
# align - ['left', 'mid', 'right']

x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(histtype='stepfilled', alpha=0.3, bins=40, range=(-10, 10))

# range - The lower and upper range of the bins. If not specified, range is (x.min(), x.max())
# density - If true, the area (or integral) under the histogram will sum to 1 and count / no. of observations * bin width
          # If not true, count / total no. of observations
          # If stacked is also true, the sum of the histograms is is normalized to 1
# stacked - If true, multiple data are stacked on top of each other
# cumulative - 
# weights -
# bottom - Location of the bottom baseline of each bin

plt.hist(x1, label='hist1', **kwargs)
plt.hist(x2, label='hist2', **kwargs)
plt.hist(x3, label='hist3', **kwargs)

plt.legend(loc='best', frameon=True)
plt.show()
```

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T

# 2-D Histograms

plt.hist2d(x, y, bins=30, cmap='Blues', alpha=0.5)
plt.colorbar()

# range - The leftmost and rightmost edges of the bins along each dimension. If not specified, [[xmin, xmax], [ymin, ymax]]
# density - Normalized histogram
# weights -
# All bins that has count less than cmin will not be displayed
# All bins that has count more than cmax will not be displayed

# Hexagonal Binning Plot

plt.figure()
plt.hexbin(x, y, gridsize=30, cmap='Blues', xscale='linear', yscale='linear', alpha=0.75, linewidths=1, edgecolors='w')
plt.colorbar()

# gridsize - No. of hexagons in x-direction and y-direction
# bins - If None, no binning is applied i.e. the color of each hexagon directly corresponds to its count value
       # If 'log' the use a logarithmic scale for the colormap
       # If an integer, divide the counts in specified no. of bins and color the hexagons accordingly
# mincnt - Only display cells with more than mincnt no. of points in the cell

plt.show()
```

```python
# Boxplots

import pandas as pd
normal_sample = np.random.normal(loc=0, scale=1, size=10000)
random_sample = np.random.random(size=10000)
gamma_sample = np.random.gamma(2, size=10000)

df = pd.DataFrame({'normal': normal_sample,
                   'random': random_sample,
                   'gamma': gamma_sample})
df.describe()
```

```python
# Points which are plotted beyond the whiskers are called fliers

plt.boxplot(df['normal'], whis=1.5, notch=True, vert=False, widths=0.55, sym=".")

# whis - Tells the boxplot to set the whisker values all the way out to the min and max values

Interquartile Range IQR=(Q3-Q1)
The upper whisker will extend to the last datum less than Q3+whis*IQR
The lower whisker will extend to the first datum greater than Q1-whis*IQR

# notch - If true, will produce a notched boxplot. These notches represent the confidence interval (CI) around the median
# vert - If false, then the boxes are drawn horizontally
# widths - Set the width of each box
# sym - Represents the default symbol for flier points

plt.figure()
plt.boxplot(df['random'], autorange=True, manage_ticks=True)
           
# autorange - If true, the data is distributed such that the 25th and 75th percentiles are equaland whis is set to 'range'
# manage_ticks - Adjust the tick locations and labels to match the boxplot positions

plt.figure()
plt.boxplot(df['gamma'], whis='range')

# whis='range' forces the whiskers to the min and max of the data

plt.figure()
plt.boxplot([df['normal'], df['random'], df['gamma']], labels=['Normal', 'Random', 'Gamma'],  meanline=True, showmeans=True, 
           meanprops=dict(ls='--', lw=1, color='g'))
plt.show()

# meanline - If true, (and showmeans is true) try to render the mean as a line spanning the full width of the box
# If you do not supply the whis parameter the whiskers only go out to halfway between the interquartile range

# showcaps, showbox, showfliers, showmeans - bool
# capprops, boxprops, whiskerprops, flierprops, medianprops, meanprops - dict
```

```python
# Contour Plots

def f(x, y):
 return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)
X, Y = np.meshgrid(x, y) # builds two-dimensional grids from one-dimensional arrays
Z = f(X, Y)

# Draw color lines:

plt.contour(X, Y, Z, levels=10, colors='black', alpha=0.75, linewidths=1)

plt.figure()
plt.contour(X, Y, Z, levels=4, colors=['r', 'y', 'g', 'b'])

# When a single color is used negative values are represented by dashed lines, and positive values by solid lines.
# levels - Detemines the no. of contour lines. If n, use n data intervals i.e. draw n+1 contour lines.

plt.figure()
plt.contour(X, Y, Z, 20, cmap='RdGy') # Adding a colomap to the plot
plt.colorbar() # Adding a colorbar to the plot

# Draw filled contours:

plt.figure()
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar()

plt.show()
```
